```python
import numpy as np
nx = 81
dx = .25
nt=51
dt = .0002   
gamma = 1.4 #cst pour l'air
L=20

#(rho,v,p) tube a t=0
IC=np.zeros((81, 3)) #là ca doit aller jusqu'à 81, sinon ca déconne
for i in range(0,40):
    IC[i]=[1.,0.,100.]
for i in range(40,81):
    IC[i]=[0.125,0.,10.]
U0=np.zeros((81, 3))
for i in range(0,81):
    U0[i]=[IC[i][0],IC[i][0]*IC[i][1],IC[i][2]/(gamma-1) + IC[i][0]*IC[i][1]**2/2]

def f(U):
    u1=U[0]
    u2=U[1]
    u3=U[2]
    test= ((u2**2)/u1) + (gamma - 1)*(u3-(0.5*(u2**2))/u1)
    testing = u3 +((gamma - 1)*(u3-(0.5*(u2**2))/u1))*u2/u1
    Fcalculee=np.array([u2,test,testing])
    return Fcalculee 

F=np.ones((81, 3))
for l in range(0,81):
    F[l]=f(U0[l])

U=U0.copy()
Udemi=np.zeros((80, 3)) #les matrices demi ont un élément de moins que les matrices U et F
for n in range(nt+1):
    Uc=U.copy()
    for j in range(0,80):
            Udemi[j]=.5*(np.array(U[j+1])+np.array(U[j]))-dt/(2*dx)*(np.array(F[j+1])-np.array(F[j]))
    for l in range(0,80):
        F[l+1]=f(Udemi[l])
    for j in range(1,81):
        U[j]=np.array(Uc[j])-dt/dx*(np.array(F[j])-np.array(F[j-1]))
    for l in range(0,81):
        F[l]=f(U[l])
        
pression=(gamma-1)*(U[50][2]-((U[50][1])**2)/2) #retourne la pression au point 2,5m (équivalent à 12.5 si on prend le début du tuyau comme originie)
print(pression)
print(U[50])
print(U)
```
